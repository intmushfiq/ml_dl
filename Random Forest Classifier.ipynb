# **Hyperparameter Tuning**

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
```

******

***Random Search CV***

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

```python
n_estimators = [int(x) for x in np.linspace(start=100,stop=2000,num=10)]        # Number of trees in random forest

max_features = [None,'sqrt','log2']                                           # Number of features to consider at every split

max_depth = [int(x) for x in np.linspace(10,1000,10)]                           # Maximum number of levels in tree

min_samples_split = [2,5,10,14]                                                 # Minimum number of samples requried to split a node

min_samples_leaf = [1,2,4,6,8]                                                  # Minimum number of samples required at each leaf node

min_impurity_decrease = [0.0, 0.1, 0.2, 0.3, 0.4]                 # A node will be split if this split induces a decrease of the impurity greater than or equal to this value

random_grid = {'n_estimators':n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_split':min_samples_split,
               'min_samples_leaf':min_samples_leaf,
               'min_impurity_decrease':min_impurity_decrease,
               'criterion':['gini', 'entropy', 'log_loss']}
```

```python
random_grid # It's a dictionary
```

Random search doesn't check all the possible combinations. But it is much faster than grid search.  
We should first use random search to get a basic idea about the approximation of desired values.

```python
rf_randomcv = RandomizedSearchCV(estimator=rfc,param_distributions=random_grid,
                           n_iter=100,cv=3,verbose=2,
                           random_state=100,n_jobs=-1)  #n_iter = 100 means the code will run for 100 iteration with 100 different combination of parameters
rf_randomcv.fit(x_train,y_train)
```

```python
rf_randomcv.best_params_
```

***Grid Search CV***

Checks all the possible combinations of hyperparameters.

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start=500,stop=1000,num=10)]        # Number of trees in random forest

max_features = ['log2']                                           # Number of features to consider at every split

max_depth = [int(x) for x in np.linspace(590,1100,10)]                           # Maximum number of levels in tree

min_samples_split = [2,3]                                                 # Minimum number of samples requried to split a node

min_samples_leaf = [1,2,3,4]                                                 # Minimum number of samples required at each leaf node

min_impurity_decrease =[0.0,0.01,0.1]                 # A node will be split if this split induces a decrease of the impurity greater than or equal to this value

random_grid = {'n_estimators':n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_split':min_samples_split,
               'min_samples_leaf':min_samples_leaf,
               'min_impurity_decrease':min_impurity_decrease,
               'criterion':['gini']}
```

```python
grid_search = GridSearchCV(estimator=rfc,param_grid=random_grid,cv=10,n_jobs=-1,verbose=2) # n_jobs = -1 -> means all the cpu cores will be used for parallel processing
                                                                                           # cv=10: This specifies the number of folds to use for cross-validation
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

# **Model**

```python
best_grid = grid_search.best_estimator_
```

```python
best_grid # model with selected hyperparameter
```

```python
y_pred = best_grid.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
print("AUC:",roc_auc_score(y_test,y_pred))
```

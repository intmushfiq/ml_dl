# **Bayesian Optimization (Automated hyperparameter tuning)**

Bayesian Optimization uses probability to find the minimum of a function. The final aim is to find the input value to a function which can give us the lowest possible output value. There are 3 main parameters in Bayesian Optimization.  
1. Objective Function - loss function to minimize  
2. Domain Space - range of input values  
3. Optimization Algorithm - search algorithm  

Bayesian Optimization builds a probability model of the objective function and uses it to select hyperparameter to evaluate in the true objective function.  

Description:  
hp -> for dyfining hyperparameter values  
STATUS_OK -> evaluation successfully done or not  
Trials -> used for minimizing objective function  
fmin -> minimize function  
TPE (Tree-structured Parzen Estimator) -> Search algorithm

```python
!pip install hyperopt
```

```python
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
```

```python
space = { 'loss': hp.choice('loss',['log_loss','exponential']), # .choice means only the value given in the square brackets will be used
          'learning_rate': hp.uniform('learning_rate',0.1,0.5),
          'n_estimators': hp.choice('n_estimators',[10,50,100,200]),
          'criterion': hp.choice('criterion',['friedman_mse','squared_error']),
          'min_samples_split': hp.choice('min_samples_split',[2,3,4]),
          'min_samples_leaf': hp.choice('min_samples_leaf',[1,2,3]),
          'max_depth': hp.choice('max_depth',[3,5,10]),
          'min_impurity_decrease': hp.uniform('min_impurity_decrease',0,0.5), # .uniform will take float values between the given two points
          'max_features': hp.choice('max_features',['sqrt','log2',None]),
          'max_leaf_nodes': hp.choice('max_leaf_nodes',[2,3,4,None]),
          'tol' : hp.uniform('tol',0.0001,0.01) # do not set it to zero
        }
```

```python
space # all hyperparameters are converted into hyperopt object
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
```

```python
def objective(space): # defining a function named objective
  GB_classifier = GradientBoostingClassifier(loss = space['loss'], # will take the loss defined in the space
                                              learning_rate = space['learning_rate'], # same for all of these
                                              n_estimators = space['n_estimators'],
                                              criterion = space['criterion'],
                                              min_samples_leaf= space['min_samples_leaf'],
                                              min_samples_split= space['min_samples_split'],
                                              max_depth= space['max_depth'],
                                              min_impurity_decrease= space['min_impurity_decrease'],
                                              max_features= space['max_features'],
                                              max_leaf_nodes= space['max_leaf_nodes'],
                                              tol = space['tol'])

  accuracy = cross_val_score(GB_classifier,x_train,y_train,cv=5).mean() # 5-fold cross validation
  # mean() retunrs the mean accuracy of the 5 validations

  # we aim to maximize accuracy, therefore, we return it as a negative value
  return{'loss': -accuracy, 'status': STATUS_OK}
```

As the loss will be minimized so the '-accuracy' will be minimized and hence the accuracy will be maximized

```python
trials = Trials() # used for minimizing objective function
best = fmin(fn = objective,         # fmin minimize the objective function fn
            space = space,
            algo = tpe.suggest,     # tpe.suggest refers to the process of suggesting the next set of hyperparameters using tpe algo
            max_evals = 80,
            trials = trials)

best # output of fmin will be stored in this variable
```

Places where we used hp.choice , we get the index as the output.  
But for hp.uniform, we get the exact value.

```python
# outcome is index often. So, we will find the value at the index by using mapping (inverse key value)

lo = {0:'log_loss',1:'exponential'} # lo - is for loss; all of these are dictionaries with index (of the parameters in space) as the key for every different hyperparameters
cri = {0:'friedman_mse',1:'squared_error'} # criterion
fea = {0:'sqrt',1:'log2',2:None} # max_features
est = {0:10, 1:50, 2:100, 3:200} # n_estimators
max_leaf = {0:2, 1:3, 2:4, 3:None} # max_leaf_nodes
min_leaf = {0:1, 1:2, 2:3} # min_sample_leaf
min_split = {0:2, 1:3, 2:4} # min_sample_split
depth = {0:3, 1:5, 2:10} # max_depth
```

```python
trainedGB = GradientBoostingClassifier(loss = lo[best['loss']],
                                      learning_rate=best['learning_rate'],
                                      n_estimators = est[best['n_estimators']],
                                      criterion=cri[best['criterion']],
                                      min_samples_split=min_split[best['min_samples_split']],
                                      min_samples_leaf=min_leaf[best['min_samples_leaf']],
                                      max_depth=depth[best['max_depth']],
                                      min_impurity_decrease=best['min_impurity_decrease'],
                                      max_features=fea[best['max_features']],
                                      max_leaf_nodes=max_leaf[best['max_leaf_nodes']],
                                      tol = best['tol'])
```

# **Test Performance**

```python
trainedGB.fit(x_train,y_train)
```

best['loss'] => 1  
lo[best['loss']] => lo[1] => 'exponential'  
loss = lo[best['loss']] => loss = 'exponential'  

similar type execution for all the nested statements

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

y_pred = trainedGB.predict(x_test)

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))

acc = accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
print("AUC:",roc_auc_score(y_test,y_pred))
```

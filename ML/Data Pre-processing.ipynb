# **Part 4: Imbalance dataset issue**

```python
df['class'].unique() #what unique values in column 'class'
```

```python
df['class'].nunique() # how many unique
```

```python
value_counts = df.groupby('class').size().reset_index(name='counts') #counting every unique values
value_counts
```

**4.A Undersampling:**
Means reducing the major class samples to match minor class samples

```python
from sklearn.utils import resample

majority_class = df[df['class'] == 1]
minority_class = df[df['class'] == 0]

n_samples = len(minority_class)
majority_downsampled = resample(majority_class, replace=False, n_samples=n_samples, random_state=42)

balanced_df = pd.concat([minority_class,majority_downsampled])
```

```python
value_counts = balanced_df.groupby('class').size().reset_index(name='counts') #counting every unique values
value_counts
```

Deleted some data from the majority class

**4.B Oversampling (SMOTE):** increasing the sample of minor class

```python
from imblearn.over_sampling import SMOTE
```

```python
X = df.drop('class',axis = 1) #dropping class column. The rest are features
y = df['class'] # Storing class in y. This is the target

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X,y)

oversampled_df = pd.DataFrame(X_resampled, columns=X.columns)
oversampled_df['class'] = y_resampled
```

```python
value_counts = oversampled_df.groupby('class').size().reset_index(name='counts') #counting every unique values
value_counts
```

# **Part 5:Outlier finding and removing**

An outlier is an anomalous value in the data that is out of range or inconsistent. Like age being 600 years. We have to remove the entire row or replace the value by mean/mode/median.

**5.A Finding from figure**

```python
import matplotlib.pyplot as plt #for plotting
import seaborn as sns
```

```python
sns.set_theme(color_codes="red")
sns.displot(data=df['Age']).set(title="Distribution of score", xlabel="scores")
```

**5.B Statistical approach**

```python
df.describe()  # describe() only works for numarical values. But since we encoded the categorical values, it will work just fine.
```

The maximum value is out of proportion. So, there is an outlier.

**5.C By finding z-score approach[-3 to 3]**

```python
z_score =(df['Age'] - df["Age"].mean())/df['Age'].std()
```

```python
for i in z_score: # value of z_score has to be between -3 to 3. For an outlier, the value would not fall in this range
  if i<-3 or i>3:
    print("You have outlier",i)
    o_val = i
  else:
    continue
```

```python
index = z_score.index[z_score == o_val][0]
index # finding the index of the outlier
```

```python
df['Age'][4]
```

```python
df["Age"][4] = df['Age'].mean() # replacing the outlier value with the mean of the column
```

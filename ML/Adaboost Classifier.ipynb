# <font color = 'orange'> <font size = 5> <b> Model

```python
from sklearn.ensemble import AdaBoostClassifier
adaboost = AdaBoostClassifier(estimator=None, n_estimators=100, learning_rate=1.0, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

k_fold = KFold(10)

accuracy = cross_val_score(adaboost,x,y,cv=k_fold,scoring='accuracy')
precision = cross_val_score(adaboost,x,y,cv=k_fold,scoring='precision')
recall = cross_val_score(adaboost,x,y,cv=k_fold,scoring='recall')
f1_score = cross_val_score(adaboost,x,y,cv=k_fold,scoring='f1')
AUC = cross_val_score(adaboost,x,y,cv=k_fold,scoring='roc_auc')
```
